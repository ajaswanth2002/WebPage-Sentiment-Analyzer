```python
# Function to count syllables in a word
def syllable_count(word):
    vowels = "aeiouAEIOU"
    count = 0
    prev_char = ""  # Initialize prev_char to an empty string

    for char in word:
        if char in vowels and prev_char not in vowels:
            count += 1
        prev_char = char

    # Handling exceptions for words ending with "es" or "ed"
    if word.endswith(('es', 'ed')):
        count -= 1

    # Handling cases where silent 'e' is present at the end
    if word.endswith('e') and not word.endswith(('le', 'ie', 'ee')):
        count -= 1

    # Minimum syllable count is 1
    return max(count, 1)

# Assuming 'final' contains the text data
words = final.split()

# Calculate syllable count for each word and sum them up
total_syllables = sum(syllable_count(word) for word in words)

# Output the total number of syllables
print("Total number of syllables:", total_syllables)
```

```python
def count_personal_pronouns(text):
    # Define a regular expression pattern to match personal pronouns
    pattern = r'\b(?:I|we|my|ours|us)\b'
    # Compile the pattern
    regex = re.compile(pattern, flags=re.IGNORECASE)
    # Find all matches in the text
    matches = regex.findall(text)
    # Return the count of matches
    return len(matches)

# Assuming 'final' contains the text data
personal_pronoun_count = count_personal_pronouns(final)

# Output the count of personal pronouns
print("Number of personal pronouns:", personal_pronoun_count)
```

```python
# Function to calculate average word length in the text
def average_word_length(text):
    # Tokenize the text into words
    words = text.split()
    # Calculate the total length of all words
    total_length = sum(len(word) for word in words)
    # Calculate the average word length
    avg_length = total_length / len(words) if words else 0
    # Return the average word length
    return avg_length

# Assuming 'final' contains the text data
avg_word_length = average_word_length(final)

# Output the average word length
print("Average word length:", avg_word_length)
```

```python
# Load the Excel file
df=pd.read_excel('/content/Output data.xlsx')

# Function to calculate sentiment scores using NLTK's Vader
def calculate_sentiment_scores_nltk(text):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(text)
    return sentiment['pos'], sentiment['neg']

# Function to calculate sentiment scores using TextBlob
def calculate_sentiment_scores_textblob(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Function to calculate average sentence length
def calculate_avg_sentence_length(text):
    sentences = nltk.sent_tokenize(text)
    words_per_sentence = [len(nltk.word_tokenize(sentence)) for sentence in sentences]
    return sum(words_per_sentence) / len(words_per_sentence)

# Function to calculate percentage of complex words
def calculate_percentage_complex_words(text):
    words = nltk.word_tokenize(text)
    tagged_words = nltk.pos_tag(words)
    complex_word_count = sum(1 for word, tag in tagged_words if tag.startswith('JJ') or tag.startswith('RB'))
    return (complex_word_count / len(words)) * 100

# Function to calculate Fog Index
def calculate_fog_index(avg_sentence_length, percentage_complex_words):
    return 0.4 * (avg_sentence_length + percentage_complex_words)

# Function to calculate average number of words per sentence
def calculate_avg_words_per_sentence(text):
    words = nltk.word_tokenize(text)
    return len(words) / len(nltk.sent_tokenize(text))

# Function to count syllables in a word
def syllable_count(text):
    vowels = "aeiouAEIOU"
    count = 0
    prev_char = ""  # Initialize prev_char to an empty string

    for char in text:
        if char in vowels and prev_char not in vowels:
            count += 1
        prev_char = char

    # Handling exceptions for words ending with "es" or "ed"
    if text.endswith(('es', 'ed')):
        count -= 1

    # Handling cases where silent 'e' is present at the end
    if text.endswith('e') and not text.endswith(('le', 'ie', 'ee')):
        count -= 1

    # Minimum syllable count is 1
    return max(count, 1)

# Function to count personal pronouns
def count_personal_pronouns(text):
    pattern = r'\b(?:I|we|my|ours|us)\b'
    regex = re.compile(pattern, flags=re.IGNORECASE)
    matches = regex.findall(text)
    return len(matches)

# Function to calculate average word length
def calculate_avg_word_length(text):
    words = text.split()
    total_length = sum(len(word) for word in words)
    return total_length / len(words) if words else 0

# Function to calculate total words in text
def total_words(text):
    words = nltk.word_tokenize(text)
    return len(words)

# Function to calculate the number of complex words
def complex_words(text):
    words = nltk.word_tokenize(text)
    tagged_words = nltk.pos_tag(words)
    complex_word_count = sum(1 for word, tag in tagged_words if tag.startswith('JJ') or tag.startswith('RB'))
    return complex_word_count
```

```python
# Apply the functions to each row of the dataframe
df['POSITIVE SCORE'], df['NEGATIVE SCORE'] = zip(*df['Extracted Text'].apply(calculate_sentiment_scores_nltk))
df['POLARITY SCORE'], df['SUBJECTIVITY SCORE'] = zip(*df['Extracted Text'].apply(calculate_sentiment_scores_textblob))
df['AVG SENTENCE LENGTH'] = df['Extracted Text'].apply(calculate_avg_sentence_length)
df['PERCENTAGE OF COMPLEX WORDS'] = df['Extracted Text'].apply(calculate_percentage_complex_words)
df['FOG INDEX'] = df.apply(lambda row: calculate_fog_index(row['AVG SENTENCE LENGTH'], row['PERCENTAGE OF COMPLEX WORDS']), axis=1)
df['AVG NUMBER OF WORDS PER SENTENCE'] = df['Extracted Text'].apply(calculate_avg_words_per_sentence)
df['COMPLEX WORD COUNT'] = df['Extracted Text'].apply(complex_words)
df['WORD COUNT'] = df['Extracted Text'].apply(total_words)
df['SYLLABLE PER WORD'] = df['Extracted Text'].apply(syllable_count)  # Ensure divisor is not zero
df['PERSONAL PRONOUNS'] = df['Extracted Text'].apply(count_personal_pronouns)
df['AVG WORD LENGTH'] = df['Extracted Text'].apply(calculate_avg_word_length)

# Save the updated dataframe back to Excel
df.to_excel('/content/drive/My Drive/Final Output File.xlsx', index=False)
```
